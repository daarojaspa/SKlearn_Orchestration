```python
import pandas as pd
path="/home/dan/PLATZI/data/SKlearn_Orchestration/skLearn_Orchestration/data/third_part/"
file="candy.csv"
df=pd.read_csv(path+file)
from sklearn.cluster import MiniBatchKMeans
X=df.drop(["competitorname"],axis=1)
kmeans=MiniBatchKMeans(n_clusters=4, batch_size=37, random_state=37).fit(X)
df["cluster"]=kmeans.predict(X)
```

Meanshift:uses a kernel like the connvolution in CNN to gader data. all the data inside  the kernell gets into a calculation of a mean. then the kernell moves to that mean and repeat the process.

```python
from sklearn.cluster import MeanShift
Ms=MeanShift().fit(X)
print(Ms.labels_)
print("#"*64)
print(Ms.cluster_centers_)
df["meanshift"]=Ms
```
